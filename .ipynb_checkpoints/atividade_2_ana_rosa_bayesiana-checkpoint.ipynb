## Exercício 1

Em uma fábrica de montagem de motores elétricos, o tempo de ciclo de uma estação de trabalho (medido em minutos e normalizado no intervalo $(0,1)$) é modelado pela seguinte densidade de probabilidade:

$$
f(x|\theta) = \theta(1-x)^{\theta - 1},\, 0<x<1,\, \theta>0
$$

O modelo reflete a realidade de que a maioria das operações é concluída rapidamente, mas há uma cauda à direita correspondente a ciclos mais longos (por falhas, ajustes manuais, etc.). O parâmetro $\theta$ controla a concentração dos tempos: quanto maior $\theta$, mais concentrados os tempos próximos de zero.

A média populacional do tempo de ciclo é dada por:

$$
\mu = E(X) = \frac{1}{\theta + 1}
$$

O engenheiro de produção deseja inferir sobre $\mu$, pois ela impacta diretamente o cálcula da capacidade produtiva da linha.


Uma amostra aleatória $X_1,...,X_n$ de tempos de ciclos normalizados foi coletada ao longo de um turno de produção.


### 1.A 

Mostre que a distribuição Gama é uma priori conjugada para $\theta$ nesse modelo e obtenha a posteriori de $\theta$.

R: Manipulando a função de densidade dada inicialmente para os dados temos

$$
f(x|\theta) = \theta(1-x)^{\theta - 1} = \theta e^{(\theta - 1) ln(1-x)}
$$

Considerando uma gama priori para $\theta$ temos:

$$
\theta \text{ ~ } Gama(\alpha, \beta)
$$

$$
\pi_{\theta}(\theta) = \frac{\beta^{\alpha}}{\Gamma(\alpha)} \theta^{\alpha - 1} e^{-\beta \theta}, \, \alpha, \beta > 0 \text{ e } \theta > 0
$$

A verossimilhança é

$$
L(x|\theta) = \prod_{i=1}^n \theta e^{(\theta - 1) ln(1-x_i)} = \theta^n e^{(\theta - 1) \sum_{i=1}^n ln(1-x_i)}
$$

E a posteriori é, então

$$
\pi_{\theta|x}(\theta) = L(x|\theta) \pi_{\theta}(\theta) = \theta^n e^{(\theta - 1) \sum_{i=1}^n ln(1-x_i)} \frac{\beta^{\alpha}}{\Gamma(\alpha)} \theta^{\alpha - 1} e^{-\beta \theta} \propto \theta^{n+\alpha - 1} e^{-\theta (\beta - \sum_{i=1}^n ln(1-x_i))}
$$

$$
\pi_{\theta|x}(\theta) \text{ ~ } Gama(n+\alpha,\beta - \sum_{i=1}^n ln(1-x_i))
$$

E, portanto, a distribuição gama é uma conjugada para $\theta$ neste modelo. Note que $\sum_{i=1}^n ln(1-x_i) < 0$, então $\beta - \sum_{i=1}^n ln(1-x_i) > 0$.

### 1.B

Derive o estimador bayesiano pontual (sob perda quadrática) para a média do tempo de ciclo $\mu = \frac{1}{\theta + 1}$. Justifique por que esse estimador não é simplesmente $\frac{1}{\hat{\theta}_{post} + 1}$, onde $\hat{\theta}_{post}$ é a média da posteriori de $\theta$.

R: Pela propriedade da média de uma $Gama(\alpha,\beta)$, para a posteriori, a média seria

$$
\hat{\theta}_{post} = E_{\pi_{\theta|x}}(\theta|x) = \frac{\alpha'}{\beta'} = \frac{n+\alpha}{\beta - \sum_{i=1}^n ln(1-x_i)}
$$

E, daí, o estimador de $\mu$ seria

$$
\hat{\mu}_{post} = \frac{1}{\hat{\theta}_{post} + 1} = \frac{1}{\frac{n+\alpha}{\beta - \sum_{i=1}^n ln(1-x_i)} + 1}
$$

Mas, neste caso, devemos derivar, pois a transformação não é linear e é convexa para $\theta > 0$, logo, $E(\frac{1}{1 + \theta}|x) \neq \frac{1}{1 + E(\theta|x)}$. Então

$$
\begin{aligned}
\hat{\mu}_{post} = E_{\pi_{\theta|x}}(\mu|x) = E_{\pi_{\theta|x}}(\frac{1}{\theta + 1}|x) \\ = \int_{0}^\infty \frac{1}{\theta + 1} \frac{(\beta - \sum_{i=1}^n ln(1-x_i))^{n+\alpha}}{\Gamma(n+\alpha)} \theta^{n+\alpha - 1} e^{-\theta (\beta - \sum_{i=1}^n ln(1-x_i))} d\theta \\ = \frac{(\beta - \sum_{i=1}^n ln(1-x_i))^{n+\alpha}}{\Gamma(n+\alpha)} \int_{0}^\infty \frac{1}{\theta + 1} \theta^{n+\alpha - 1} e^{-\theta (\beta - \sum_{i=1}^n ln(1-x_i))} d\theta
\end{aligned}
$$

Considerando a integral fundamental, assumindo as condições de convergência $\mathfrak{R}(a) > 0$ e $\mathfrak{R}(s) > 0$ temos

$$
I(s,a) = \int_{0}^\infty \frac{\theta^{s-1} e^{-a\theta}}{\theta + 1} d\theta = e^a a^{-s} \Gamma(s) \Gamma(s, a)
$$

Onde $\Gamma(s, a) = \int_{a}^\infty t^{s-1} e^{-t} dt = \Gamma(s) - \int_{0}^a t^{s-1} e^{-t} dt, \, a>0 $ é a função gama incompleta superior.

Logo, aplicando ao problema temos

$$
\begin{aligned}
\hat{\mu}_{post} = \frac{(\beta - \sum_{i=1}^n ln(1-x_i))^{n+\alpha}}{\Gamma(n+\alpha)} \int_{0}^\infty \frac{1}{\theta + 1} \theta^{n+\alpha - 1} e^{-\theta (\beta - \sum_{i=1}^n ln(1-x_i))} d\theta \\ = \frac{(\beta - \sum_{i=1}^n ln(1-x_i))^{n+\alpha}}{\Gamma(n+\alpha)} e^{\beta - \sum_{i=1}^n ln(1-x_i)} \Gamma(n+\alpha) \Gamma(n+\alpha, \beta - \sum_{i=1}^n ln(1-x_i)) \\ = \beta - \sum_{i=1}^n ln(1-x_i))^{n+\alpha} e^{\beta - \sum_{i=1}^n ln(1-x_i)} \Gamma(n+\alpha, \beta - \sum_{i=1}^n ln(1-x_i))
\end{aligned}
$$

O que vale pois $\mathfrak{R}(\beta - \sum_{i=1}^n ln(1-x_i)) > 0$ e $\mathfrak{R}(n+\alpha) = n+\alpha > 0$, pois $n > 0$ e $\alpha > 0$.

Logo, o estimador bayesiano de $\mu$ sob perda quadrática é

$$
\hat{\mu}_{post} = \beta - \sum_{i=1}^n ln(1-x_i))^{n+\alpha} e^{\beta - \sum_{i=1}^n ln(1-x_i)} \Gamma(n+\alpha, \beta - \sum_{i=1}^n ln(1-x_i))
$$


### 1.C

Construa um intervalo de credibilidade de 95% para $\mu$. Explique como você o obteria na prática (analiticamente ou numericamente).

R: Usando a posteriori de $\theta$, podemos criar intervalo de credibilidade para $\mu$ usando a distribuição acumulada de $\theta|x$ e $\mathcal{U} = \frac{1}{\theta + 1}$

$$
P(\mathcal{U} \leq \mu) = P(\frac{1}{\theta + 1} \leq \mu) = P({\theta + 1} > \frac{1}{\mu}) = 1 - P(\theta \leq \frac{1}{\mu} - 1) ,\,0 < \mu < 1
$$

Como a acumulada de $\theta|x$ é

$$
\begin{aligned}
F_{\theta|x}(\theta) = P(\theta \leq \theta) = \int_{0}^{\theta} \frac{(\beta - \sum_{i=1}^n ln(1-x_i))^{n+\alpha}}{\Gamma(n+\alpha)} t^{n+\alpha - 1} e^{-t (\beta - \sum_{i=1}^n ln(1-x_i))} d t\\ = F_{gama \,inc}(\theta)
\end{aligned}
$$

E, portanto,

$$
P(\mathcal{U} \leq \mu) = 1 - P(\theta \leq \frac{1}{\mu} - 1) = 1 - F_{gama \,inc}(\frac{1}{\mu} - 1)
$$

Para $\theta$, o intervalo de credibilidade de 95% igual-caudal é

$$
IC_{\theta} = \int_{0,025}^{0,975} \frac{(\beta - \sum_{i=1}^n ln(1-x_i))^{n+\alpha}}{\Gamma(n+\alpha)} \theta^{n+\alpha - 1} e^{-\theta (\beta - \sum_{i=1}^n ln(1-x_i))} d \theta
$$

Que são os quantis 2,5% e 97,5% da posteriori: $q_{0,025},\,q_{0,975}$, logo o IC de 95% para $\mu$ é 

$$
P(q_{0,025} \leq \theta \leq q_{0,975}) = 0,95 
$$
$$
P(q_{0,025} +1 \leq \theta +1 \leq q_{0,975}+1) = 0,95 
$$
$$
P(\frac{1}{q_{0,025} +1} > \frac{1}{1+\theta} > \frac{1}{q_{0,975} +1} ) = 0,95 
$$

$$
P(\frac{1}{q_{0,975} +1} \leq \mu \leq \frac{1}{q_{0,025} +1}) = 0,95 
$$

$$
IC_{\mu} = [\frac{1}{1 + q_{0,975}},\frac{1}{1 + q_{0,025}}]
$$

Esta integral só pode ser resolvida numericamente, então usamos algoritmos para obter os intervalos.

Considere n = 25 observações simuladas com $\theta = 2 (logo, \mu = \frac{1}{3})$. Use uma priori Gamma(2,1). Implemente em R ou Python

R: Para obter amostras da distribuição, usaremos o fato da distribuição ser uma $Beta(1,\theta)$.

```python
# Gerando 25 elementos de theta(1-x)^(theta-1) com parametro theta com priori beta

alpha = 2
beta = 1
theta = 2
n = 25

amostras_x = st.beta.rvs(a = 1, b = theta, size = n, random_state=1234)
amostras_x
```

$$
\begin{aligned}
\hat{\mu}_{post} = E_{\pi_{\theta|x}}(\mu|x) = E_{\pi_{\theta|x}}(\frac{1}{\theta + 1}|x)\\ = \int_{0}^\infty \frac{1}{\theta + 1} \frac{(\beta - \sum_{i=1}^n ln(1-x_i))^{n+\alpha}}{\Gamma(n+\alpha)} \theta^{n+\alpha - 1} e^{-\theta (\beta - \sum_{i=1}^n ln(1-x_i))} d\theta\\ = \frac{(\beta - \sum_{i=1}^n ln(1-x_i))^{n+\alpha}}{\Gamma(n+\alpha)} \int_{0}^\infty \frac{1}{\theta + 1} \theta^{n+\alpha - 1} e^{-\theta (\beta - \sum_{i=1}^n ln(1-x_i))} d\theta
\end{aligned}
$$

```python
# Obtendo a estimativa via reamostragem ponderada

def importance_sampling_prior_aux(x, a, b, n_samples=25):
    """
    Calcula E[1/(1+theta)|x] via reamostragem ponderada
    usando a priori Gamma(a,b) como distribuição auxiliar.
    """
    x = np.array(x)
    n = len(x)
    s = np.sum(np.log(1 - x))
    
    # Amostragem da distribuição auxiliar (a priori)
    thetas = np.random.gamma(shape=a, scale=1/b, size=n_samples)

    # Pesos de importância (sem constante normalizadora)
    log_w = n * np.log(thetas) + s * thetas
    w = np.exp(log_w - np.max(log_w))  # normalização numérica segura
    
    # Quantidade de interesse
    h = 1 / (1 + thetas)

    # Estimativa por amostragem ponderada
    estimate = np.sum(h * w) / np.sum(w)
    return estimate
```

```python
mu_post = importance_sampling_prior_aux(amostras_x, a=alpha, b=beta, n_samples=25)

print(f'Estimador Bayesiano Pontual de mu é: {mu_post}')
```

```python
# Intervalo de credibilidade de 95%

# Para theta

s = sum([np.log(1-i) for i in amostras_x])
alpha_post = alpha + n
beta_post = beta - s

q1_theta = st.gamma.ppf(q = 0.025, a = alpha_post, scale = 1/beta_post)
q2_theta = st.gamma.ppf(q = 0.975, a = alpha_post, scale = 1/beta_post)

ic_theta = [q1_theta,q2_theta]
print(f'IC igual-caudal de theta de 95%: {q1_theta} a {q2_theta}\n')

# Para mu (podemos fazer direto como 1/1+theta)

q1_mu = 1/(1+ q2_theta)
q2_mu = 1/(1+ q1_theta)

ic_theta = [q1_mu,q2_mu]
print(f'IC igual-caudal de 95% de mu a partir de theta é: {q1_mu} a {q2_mu}\n')
```

## Exercício 2

Em uma campanha de marketing digital, o número de cliques ou interações que um cliente realiza antes de realizar uma conversão (ex: compra, inscrição) é modelado como uma variável aleatória X com distribuição geométrica:

$$
P(X = x|\theta) = \theta(1-\theta)^x,\, x = 0,1,2,...\, e\, 0 < \theta < 1
$$

Onde $\theta$ representa a probabilidade de conversão em cada interação. O parâmetro de interesse para a equipe de marketing é a probabilidade não conversão por interação, que é dada por:

$$
\gamma = g(\theta) = 1 - \theta
$$

Uma amostra aleatória $X_1, ..., X_n$ foi coletada ao longo de uma semana, registrando o número de interações até a conversão para $n$ clientes distintos.

### 2.A

Mostre que a distribuição Beta é uma priori conjugada para este modelo.

R: Assumindo $\theta \text{ ~ }Beta(\alpha, \beta)$ temos

$$
\pi(\theta) = \frac{\theta^{\alpha - 1} (1-\theta)^{\beta - 1}}{B(\alpha,\beta)}, \, B(\alpha, \beta) = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha + \beta)}
$$

A verossimilhança dos dados é

$$
L(x|\theta) = \prod_{i=1}^n \theta(1-\theta)^{x_i} = \theta^n (1-\theta)^{\sum_{i=1}^n x_i}
$$

Logo, a posteriori de $\theta$ é

$$
\pi_{\theta|x}(\theta) \propto L(x|\theta)\pi(\theta) \propto \theta^n (1-\theta)^{\sum_{i=1}^n x_i}\frac{\theta^{\alpha - 1} (1-\theta)^{\beta - 1}}{B(\alpha,\beta)} \propto \underbrace{\theta^{\alpha + n - 1} (1-\theta)^{\beta + \sum_{i=1}^n x_i - 1}}_{\text{Núcleo de }Beta(\alpha + n, \beta+\sum_{i=1}^n x_i )}
$$

Portanto, a distribuição Beta é uma priori conjugada para o modelo geométrico.

### 2.B

Derive o estimador bayesiano de $\gamma = 1 - \theta$ sob perda quadrática.

R: Temos que o estimador bayesiano de $\gamma$ é

$$
\begin{aligned}
E_{\theta|x}(1-\theta|x) = \frac{1}{B(\alpha +n, \beta + \sum_{i=1}^n x_i)}\int_{0}^1 (1-\theta)\theta^{\alpha + n - 1} (1-\theta)^{\beta + \sum_{i=1}^n x_i - 1} d\theta \\= \frac{1}{B(\alpha +n, \beta + \sum_{i=1}^n x_i)}\underbrace{\int_{0}^1 (\theta^{\alpha + n - 1} (1-\theta)^{\beta + \sum_{i=1}^n x_i + 1 - 1} d\theta}_{\text{Núcleo de }Beta(\alpha + n, \beta + \sum_{i=1}^n x_i + 1)}\\ = \frac{B(\alpha +n, \beta + \sum_{i=1}^n x_i + 1)}{B(\alpha +n, \beta + \sum_{i=1}^n x_i)} \int_{0}^1 \frac{(\theta^{\alpha + n - 1} (1-\theta)^{\beta + \sum_{i=1}^n x_i + 1 - 1}}{B(\alpha +n, \beta + \sum_{i=1}^n x_i + 1)} d\theta \\= \frac{B(\alpha +n, \beta + \sum_{i=1}^n x_i + 1)}{B(\alpha +n, \beta + \sum_{i=1}^n x_i)}\\ = \frac{\Gamma(\alpha +n)\Gamma(\beta + \sum_{i=1}^n x_i + 1)\,\Gamma(\alpha +n+ \beta + \sum_{i=1}^n x_i)}{\Gamma(\alpha +n+ \beta + \sum_{i=1}^n x_i + 1)\, \Gamma(\alpha +n)\Gamma(\beta + \sum_{i=1}^n x_i))}\\ = \frac{\Gamma(\beta + \sum_{i=1}^n x_i + 1)\,\Gamma(\alpha +n+ \beta + \sum_{i=1}^n x_i)}{\Gamma(\alpha +n+ \beta + \sum_{i=1}^n x_i + 1)\, \Gamma(\beta + \sum_{i=1}^n x_i))}
\end{aligned}
$$

Como $\Gamma(\alpha +n+ \beta + \sum_{i=1}^n x_i + 1) = (\alpha +n+ \beta + \sum_{i=1}^n x_i)\Gamma(\alpha +n+ \beta + \sum_{i=1}^n x_i)$ e $\Gamma(\beta + \sum_{i=1}^n x_i + 1) = (\beta + \sum_{i=1}^n x_i)\Gamma(\beta + \sum_{i=1}^n x_i)$ temos

$$
\hat{\gamma}_{post} = E_{\theta|x}(1-\theta|x) = \frac{(\beta + \sum_{i=1}^n x_i)}{(\alpha +n+ \beta + \sum_{i=1}^n x_i)}
$$

Que é obtido analiticamente.

### 2.C

Construa um intervalo de credibilidade de 95% para $\gamma$.

R: Um intervalo de credibilidade de 95% igual-caudal para $\theta$ é 

$$
\int_{0,025}^{0,975} \frac{1}{B(\alpha +n, \beta + \sum_{i=1}^n x_i)}\theta^{\alpha + n - 1} (1-\theta)^{\beta + \sum_{i=1}^n x_i - 1} d\theta
$$

Que são os quantis 0,025 ($q_{0,025}$) e 0,975 ($q_{0,975}$) da $Beta(\alpha + n, \beta+\sum_{i=1}^n x_i)$. Então, para $\gamma$:

$$
IC_{\gamma} = [1 - q_{0,975};1-q_{0,025}]
$$

Ou, então, se $\gamma = 1 - \theta$ e $\theta $ ~ $Beta(\alpha, \beta)$, então $\gamma$ ~$Beta(\beta, \alpha) e o IC é

$$
\int_{0,025}^{0,975} \frac{1}{B(\beta + \sum_{i=1}^n x_i,\alpha +n)}(1 - \gamma)^{\alpha + n - 1} \gamma^{\beta + \sum_{i=1}^n x_i - 1} d\gamma
$$

E, daí, buscamos os quantis 2,5% e 97,5%.

Se for o intervalo de credibilidade HPD, precisamos obter numericamente via algoritmos.

### 2.D

A equipe quer testar se a probabilidade de não conversão por interação é igual a 0,7, ou seja, $\gamma = 0,7$. Formule e conduza um teste de hipoteses bayesiano bilateral:

$$
H_0: \gamma = 0,7 \text{ vs } H_1:\gamma \neq 0,7
$$

Explique como você usaria a densidade a posteriori de $\gamma$ para calcular a probabilidade a posteriori de $H_0$ (por exemplo, via aproximação de região de alta densidade ou fator de bayes com uma priori mista).

R: Neste caso, dá para utilizar os intervalos de confiança, tanto para $\gamma$ quanto para $\theta$, focando no HPD, que possui o menor intervalo de valores que contém ($1 - \alpha$)% de massa. Se o intervalo possui o valor a ser testado, temos evidência a favor de $H_0$.

Do mesmo modo, podemos utilizar priori mista com fator de Bayes para adicionar massa à probabilidade de $H_0$ e testar a hipótese. No caso, usaremos $\theta$ para realizar o teste:

$$
H_0: \theta = 0,3 \text{ vs }H_1:\theta \neq 0,3
$$

A densidade marginal sob $H_0$ é
$$
P(x|H_0) = P(x|\theta = 0,3) = (0,3)^n(1 - 0,3)^{\sum_{i=1}^n x_i}
$$

A evidência marginal sob $H_1$ é calculada integrando a verossimilhança com a priori de $\theta$ (que é $Beta(\alpha, \beta)$):

$$
\begin{aligned}
P(x|H_1) = \int_{0}^1 \theta^n (1-\theta)^{\sum_{i=1}^n x_i} \frac{\theta^{\alpha - 1} (1-\theta)^{\beta - 1} d\theta}{B(\alpha, \beta)} \\= \int_{0}^1 \frac{\theta^{\alpha + n - 1} (1-\theta)^{\beta + \sum_{i=1}^n x_i - 1} d\theta}{B(\alpha, \beta)} = \frac{1}{B(\alpha, \beta)} \int_{0}^1 \underbrace{\theta^{\alpha + n - 1} (1-\theta)^{\beta + \sum_{i=1}^n x_i - 1} d\theta}_{\text{Núcleo de }Beta(\alpha + n, \beta + \sum_{i=1}^n x_i)}\\ = \frac{B(\alpha + n, \beta + \sum_{i=1}^n x_i)}{B(\alpha, \beta)} \int_{0}^1 \frac{\theta^{\alpha + n - 1} (1-\theta)^{\beta + \sum_{i=1}^n x_i - 1} d\theta}{B(\alpha + n, \beta + \sum_{i=1}^n x_i)}\\ = \frac{B(\alpha + n, \beta + \sum_{i=1}^n x_i)}{B(\alpha, \beta)}
\end{aligned}
$$

Logo, para testar $H_0:\theta = 0,3$ vs $H_1: \theta \neq 0,3$ usamos o fator de Bayes:

$$
B = \frac{P(x|H_0)}{P(x|H_1)} = \frac{(0,3)^n(1 - 0,3)^{\sum_{i=1}^n x_i}}{\frac{B(\alpha + n, \beta + \sum_{i=1}^n x_i)}{B(\alpha, \beta)}}
$$

E, quanto maior o fator, maior evidência para $H_0$.

### 3.A (Aproximação de Laplace)

Use a função laplace(R) para obter a moda a posteriori ($\tilde{\theta},\tilde{\sigma}$) e a matriz de covariância aproximada $V^{-1}$. Discuta: a posteriori é unimodal? A aproximação normal é adequada?

```python
def log_verossimilhanca(x, theta, sigma,n):
    if sigma <= 0:
        return -np.inf
    else:
        z = (x - theta)/sigma
        s = sum(z)
        s2 = sum([np.log(1 + np.exp(-i)) for i in z])
        log_veros = -n*np.log(sigma) - s - 2*s2
        return log_veros

def log_priori(theta,sigma):
    lp_theta = -np.log(2*math.pi)/2 - np.log(100)/2 - theta**2/20
    lp_sigma = np.log(2) - np.log(5) - np.log(math.pi) - np.log(1 + (sigma/5)**2)
    return lp_theta + lp_sigma

def log_posteriori(x, theta, sigma):
    n = len(x)
    lp_vero = log_verossimilhanca(x, theta, sigma,n)
    lp_priori = log_priori(theta,sigma)
    return lp_vero + lp_priori

# vamos transformar sigma para que sigma > 0
def log_posteriori_transform(params, x):
    theta, log_sigma = params
    sigma = np.exp(log_sigma)
    return log_posteriori(x, theta, sigma)
```

```python
# Aproximação via Laplace

# 1 - Função negativa da log-posteriori (para minimizar)
def neg_logpost(params, x):
    return -log_posteriori_transform(params, x)

# 2 - Chute inicial
x0 = [np.mean(x), np.log(np.std(x))]

# 3 - Encontrando MAP
res = minimize(neg_logpost, x0, args=(x,), method='BFGS')
theta_map, log_sigma_map = res.x
sigma_map = np.exp(log_sigma_map)
print(f'Estimativa de theta (MAP): {theta_map}')
print(f'Estimativa de log_sigma (MAP): {log_sigma_map}')
print(f'Estimativa de sigma (MAP): {np.exp(log_sigma_map)}')

# 4 - Hessiana da negativa log-posteriori no MAP
H = nd.Hessian(lambda p: neg_logpost(p, x))(res.x)

# 5 - Covariancia da normal aproximada
V = np.linalg.inv(H)
print(f'Matriz de covariância aproximada:\n{V}')

print('Aproximação de Laplace:')
print(f'(theta, log(sigma)) ~ N_2(({theta_map}, {log_sigma_map}), V)')
print('Normal bivariada')
```

```python
# --- 1) Criar grade de valores de theta e sigma
theta_vals = np.linspace(theta_map - 3*np.sqrt(V[0,0]), theta_map + 3*np.sqrt(V[0,0]), 50)
sigma_vals = np.linspace(max(0.01, sigma_map - 3*np.sqrt(V[1,1])), sigma_map + 3*np.sqrt(V[1,1]), 50)
Theta, Sigma = np.meshgrid(theta_vals, sigma_vals)

# --- 2) Avaliar posteriori real na grade
Posteriori_real = np.zeros_like(Theta)
for i in range(Theta.shape[0]):
    for j in range(Theta.shape[1]):
        Posteriori_real[i,j] = np.exp(log_posteriori(x, Theta[i,j], Sigma[i,j]))

# --- 3) Avaliar posteriori pela aproximação de Laplace (Normal em log_sigma)
from scipy.stats import multivariate_normal

# grade em (theta, log(sigma))
LogSigma = np.log(Sigma)
pos = np.dstack((Theta, LogSigma))
rv = multivariate_normal([theta_map, log_sigma_map], V)
Posteriori_laplace = rv.pdf(pos)

# --- 4) Plot
fig = plt.figure(figsize=(12,5))

# Posteriori real
ax1 = fig.add_subplot(1,2,1, projection='3d')
ax1.plot_surface(Theta, Sigma, Posteriori_real, cmap='viridis')
ax1.set_title('Posteriori real')
ax1.set_xlabel('theta')
ax1.set_ylabel('sigma')

# Aproximação Laplace
ax2 = fig.add_subplot(1,2,2, projection='3d')
ax2.plot_surface(Theta, Sigma, Posteriori_laplace, cmap='plasma')
ax2.set_title('Aproximação de Laplace')
ax2.set_xlabel('theta')
ax2.set_ylabel('sigma')

plt.tight_layout()
plt.show()
```

### 3.D (Metropolis-Hastings) Caminhada Aleatória 

Use rwmetrop com vetor de parâmetros ϑ = (θ,log σ) (para garantir σ > 0). Gere 20.000 iterações, aplique burn-in e thinning. Apresente diagnósticos (traço, ACF, densidade marginal).

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
import seaborn as sns

# --- Defina o MCMC Random Walk Metropolis
def rw_metropolis(n_iter, prop_cov, x, posteriori_logsigma):
    aceitos = 0
    samples = np.zeros((n_iter, 2))
    # Estado inicial (usei MAP como ponto de partida)
    theta_curr, log_sigma_curr = theta_map, log_sigma_map  
    lp_curr = np.log(posteriori_logsigma([theta_curr, log_sigma_curr]))
    for t in range(n_iter):
        theta_prop, log_sigma_prop = np.random.multivariate_normal(
            [theta_curr, log_sigma_curr], prop_cov
        )
        lp_prop = np.log(posteriori_logsigma([theta_prop, log_sigma_prop]))
        # aceitação
        alpha = np.exp(lp_prop - lp_curr)
        if np.random.rand() < min(1, alpha):
            theta_curr, log_sigma_curr = theta_prop, log_sigma_prop
            lp_curr = lp_prop
            aceitos += 1
        samples[t] = [theta_curr, log_sigma_curr]
    return samples, aceitos/n_iter

# --- Parâmetros do MCMC
n_iter = 20000
prop_cov = np.array([[0.1, 0.0],
                     [0.0, 0.1]])  # você pode ajustar a escala para boa aceitação

# --- Rodar
samples, tx_aceitacao_metrop = rw_metropolis(n_iter, prop_cov, x, posteriori_logsigma)

# --- Burn-in e thinning
burnin = 5000
thin = 5
samples_post = samples[burnin::thin]
theta_chain = samples_post[:, 0]
sigma_chain = np.exp(samples_post[:, 1])
```

### 3.E (Metropolis-within-Gibbs)

Esboce um algoritmo Gibbs onde:
- θ | σ,x: atualizado por Metropolis (não conjugado);
- σ | θ,x: atualizado por Metropolis (não conjugado). Implemente e compare com o Metropolis bivariado.

```python
def metropolis_within_gibbs(n_iter, x, init_theta, init_logsigma, tau_theta, tau_logsigma):
    aceitos = 0
    samples = np.zeros((n_iter, 2))  # coluna0 = theta, coluna1 = log_sigma
    theta_curr = init_theta
    logsigma_curr = init_logsigma

    lp_curr = log_posteriori(x, theta_curr, np.exp(logsigma_curr))

    for t in range(n_iter):
        # ---- Atualiza theta dado sigma ----
        theta_prop = np.random.normal(theta_curr, tau_theta)
        lp_theta_prop = log_posteriori(x, theta_prop, np.exp(logsigma_curr))
        # razão (porque σ fixo, comparar posteriors completos funciona)
        alpha_theta = np.exp(lp_theta_prop - lp_curr)
        if np.random.rand() < min(1, alpha_theta):
            theta_curr = theta_prop
            lp_curr = lp_theta_prop
            aceitos += 1

        # ---- Atualiza log(sigma) dado theta ----
        logsigma_prop = np.random.normal(logsigma_curr, tau_logsigma)
        sigma_prop = np.exp(logsigma_prop)
        lp_logsigma_prop = log_posteriori(x, theta_curr, sigma_prop)
        alpha_logsigma = np.exp(lp_logsigma_prop - lp_curr)
        if np.random.rand() < min(1, alpha_logsigma):
            logsigma_curr = logsigma_prop
            lp_curr = lp_logsigma_prop

        samples[t, :] = [theta_curr, logsigma_curr]

    return samples, aceitos/n_iter

# Parâmetros para rodar
n_iter = 20000
init_theta = theta_map
init_logsigma = log_sigma_map
tau_theta = 0.1         # desvio padrão da proposta para theta
tau_logsigma = 0.1      # desvio para log sigma

samples_gibbs, tx_aceitacao_gibbs = metropolis_within_gibbs(n_iter, x, init_theta, init_logsigma, tau_theta, tau_logsigma)

# Burn-in e thinning
# Descartamos as 5 mil primeiras amostras
burnin = 5000

# Considerando 1 valor amostrado a cada 5 simulações
thin = 5
samples_gibbs_thinned = samples_gibbs[burnin::thin]
theta_gibbs = samples_gibbs_thinned[:, 0]
sigma_gibbs = np.exp(samples_gibbs_thinned[:, 1])
```

### 3.G Comparação Final

Construa uma tabela comparando todos os métodos em termos de:
- estimativas de θ e σ;
- eficiência (ESS/segundo, se possível);
- facilidade de implementação;
- robustez a escolhas iniciais.

```python
# ---------------------------------------------------------
# Effective Sample Size (ESS) baseado na soma das autocorrelações
# ---------------------------------------------------------
def ess_mcmc(samples, max_lag=None):
    samples = np.asarray(samples)
    n = len(samples)
    
    if max_lag is None:
        # Regra comum: usar até n/2 ou até onde autocorrelação fica negativa
        max_lag = n // 2
    
    # Autocovariância com média removida
    mean = np.mean(samples)
    var = np.var(samples, ddof=1)
    
    def autocorr(lag):
        return np.sum((samples[:n-lag] - mean) * (samples[lag:] - mean)) / ((n-lag) * var)

    rho = []
    for lag in range(1, max_lag):
        ac = autocorr(lag)
        if ac < 0:   # parar quando autocorrelação fica negativa
            break
        rho.append(ac)

    # ESS formula
    ess = n / (1 + 2 * np.sum(rho))
    return ess

# ---------------------------------------------------------
# Erro Quadrático Médio (MSE)
# ---------------------------------------------------------
def eqm(estimativas, valor_real):
    estimativas = np.asarray(estimativas)
    return np.mean((estimativas - valor_real) ** 2)
```
