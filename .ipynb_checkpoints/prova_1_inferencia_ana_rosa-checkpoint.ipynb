```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

## Definindo f.d.p.s usadas no código

Uma v.a. X tem distribuição Beta se sua f.d.p. é dada por

$$
f(x;\alpha,\beta) = \begin{cases}
\frac{1}{B(\alpha,\beta)} x^{\alpha-1} (1-x)^{\beta-1}, \text{ se } 0 \leq x \leq 1, \, \alpha>0, \, \beta>0 \\
0, \text{ caso contrário}
\end{cases}
$$

Onde
$$
B(\alpha,\beta) = \frac{\Gamma(\alpha).\Gamma(\beta)}{\Gamma(\alpha + \beta)}
$$

$$
\Gamma(\alpha) = \int_{0}^{\infty} e^{-x} x^{\alpha - 1} dx, \text{ com }\alpha > 0
$$

## Esperança

Temos
$$
E(X) = \frac{\alpha}{\alpha + \beta}
$$

## Variância
Temos
 
$$
Var(X) = \frac{\alpha \beta}{(\alpha+\beta)^2 (\alpha + \beta + 1)}
$$

```python
def gama(a):
    return(math.factorial(a-1))

def beta(a,b):
    return gama(a)*gama(b)/gama(a+b)


def dist_beta(a,b):
    suporte = np.linspace(0,1,100)
    soma, p = 0, 0
    p_x, p_acum = [], []
    
    for i in suporte:
        p = (1/(beta(a,b)))*i**(a-1)*(1-i)**(b-1)
        soma = soma + p
        p_x.append(p)
        p_acum.append(soma)
    
    de = pd.DataFrame({
            'x':suporte, 
            'p_x': p_x, 
            'p_acum': p_acum
        })

    return de
```

<p>A distribuição uniforme é a v.a. onde cada evento tem a mesma probabilidade de ocorrer. Sua f.d.p. (função de densidade de probabilidade) é dada por</p>

$$
f(x; \alpha, \beta) = \begin{cases}
 \frac{1}{\beta - \alpha}, \, \alpha \leq x \leq \beta \\
 0, \text{ caso contrário }
\end{cases}
$$

## Esperança

$$
E(x) = \frac{\beta + \alpha}{2}
$$

## Variância

$$
Var(x) = \frac{(\beta - \alpha)^2}{12}
$$

## Notação:
X ~ Un($\alpha$, $\beta$)

```python
def dist_uniforme_continua(a,b):
    suporte = np.linspace(a,b,100)
    soma, p = 0, 0
    p_x, p_acum = [], []
    
    for i in suporte:
        if i >= a and i <= b:
            p = 1/(b-a) 
        else:
            p = 0
        soma = soma + p
        p_x.append(p)
        p_acum.append(soma)
    
    du = pd.DataFrame({
            'x':suporte, 
            'p_x': p_x, 
            'p_acum': p_acum
        })

    return du
```

As funções em Python são scipy.special.digamma e scipy.special.polygamma (com n=1)

```python
# O código em R é (passando para Python)

from scipy.special import digamma, polygamma

# Função U
def u(theta,x):
    alpha = theta[0]
    beta = theta[1]
    n = len(x)
    #Obtém o valor vetorial da função U no ponto (alpha, beta)
    U1 = digamma(alpha) - digamma(alpha+beta) - sum([math.log(i) for i in x])/n
    U2 = digamma(beta) - digamma(alpha+beta) - sum([math.log(1-i) for i in x])/n
    return [U1,U2]

# Função H
def h(theta,x):
    alpha = theta[0]
    beta = theta[1]
    #Calcula matriz H no ponto (alpha, beta)
    h11 = polygamma(1,alpha)-polygamma(1,alpha+beta)
    h12 = polygamma(1,alpha+beta)
    h21 = polygamma(1,alpha+beta)
    h22 = polygamma(1,beta)-polygamma(1,alpha+beta)
    return np.array([[h11,h12],[h21,h22]])

# Função H_inv (np.linalg.inv(A))
def h_inv(theta,x):
    #Calcula a matriz inversa de H no ponto (alpha, beta)
    hi = h(theta,x)
    return np.linalg.inv(hi)

def maxima_verossimilhanca_beta(x, niter=10):
    #Chutes iniciais: theta0 = [alpha0, beta0]
    theta0 = [1,1]
    #Lista de vetores estimados ao longo das iterações
    thetas = [theta0]
    k = 1
    # utilizando s para comparar com erro
    s = [1,1]
    
    #iterações do algoritmo de Newton-Raphson
    #for k in range(1,niter+1):
    while abs(s[0]) > 10**(-4) and abs(s[1]) > 10**(-4):
        m1 = np.linalg.inv(h(thetas[k-1],x))
        m2 = u(thetas[k-1],x)
        print(f'm1: {m1}')
        print(f'm2: {m2}')
        s = np.dot(m1,m2)
        theta_k = thetas[k-1] - s
        print(f'primeira parte: {thetas[k-1]}')
        print(f'segunda parte: {s}')
        print(f'resultado: {theta_k}')
        thetas.append(theta_k)
        print(theta_k)
                
        k += 1

    return thetas
```

```python
# gerando 5000 amostras de tamanho n da distribuicao beta(3,5)
# n = 10 inicio
# n = 1000 ....

n = 10
amostras = []
eixo_x, eixo_y = [], []
for i in range(5000):
    x = list(beta.rvs(3,5,size=n))
    amostras.append(x)
    a = 3
    # dado a, qual b que maximiza?
    b = a*(1-(sum(x)/len(x)))/(sum(x)/len(x))
    eixo_y.append(b)
    
    b = 5
    # dado b, qual a que maximiza?
    a = b*(sum(x)/len(x))/(1-(sum(x)/len(x)))
    eixo_x.append(a)
```
